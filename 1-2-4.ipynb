```python
CH1= Dataset('OR_ABI-L1b-RadF-M6C01_G16_s20202621930205_e20202621939513_c20202621939571.nc')
CH2 = Dataset('OR_ABI-L1b-RadF-M6C02_G16_s20202621930205_e20202621939513_c20202621939561.nc')
CH4= Dataset('OR_ABI-L1b-RadF-M6C04_G16_s20202621930205_e20202621939513_c20202621939546.nc')

```

```python
CH1

np.shape(CH1.variables['Rad'][:])

```

```python
np.shape(CH2.variables['Rad'][:])
```

```python

np.shape(CH4.variables['Rad'][:])
```

```python
BLUE = CH1.variables['Rad'][:][::2,::2]
GREEN =CH4.variables['Rad'][:][::1,::1]
RED=CH2.variables['Rad'][:][::4,::4]

RED=(RED-RED.min())/(RED.max()-RED.min())
GREEN=(GREEN-GREEN.min())/(GREEN.max()-GREEN.min())
BLUE=(BLUE-BLUE.min())/(BLUE.max()-BLUE.min())
```

```python
np.shape(RED)
```

```python
np.shape(GREEN)
```

```python
np.shape(BLUE)
```

```python

COLORES=np.zeros((5424, 5424,3))

COLORES[:,:,0]=RED

COLORES[:,:,1]=GREEN

COLORES[:,:,2]=BLUE

plt.figure(figsize = (5,5),dpi=150)
plt.imshow(COLORES)

plt.figure(figsize = (5,5),dpi=150)
plt.xlim(2000,2500)
plt.ylim(2300,2000)
plt.title('RGB con canales 1-2-4 a las 19:30 UTC con el producto "raciances"')
plt.imshow(COLORES)

plt.figure(figsize = (5,5),dpi=150)
plt.xlim(2200,2300)
plt.ylim(2300,2200)
plt.title('RGB con canales 1-2-4 a las 19:30 UTC con el producto "raciances"')
plt.imshow(COLORES)
```

```python
CH1= Dataset('OR_ABI-L1b-RadF-M6C01_G16_s20202621930205_e20202621939513_c20202621939571.nc')
CH2 = Dataset('OR_ABI-L1b-RadF-M6C02_G16_s20202621930205_e20202621939513_c20202621939561.nc')
CH3= Dataset('OR_ABI-L1b-RadF-M6C03_G16_s20202621930205_e20202621939513_c20202621939577.nc')
```

```python
CH1
```

```python
np.shape(CH1.variables['Rad'][:])
```

```python
np.shape(CH2.variables['Rad'][:])
```

```python
np.shape(CH3.variables['Rad'][:])
```

```python
BLUE = CH1.variables['Rad'][:][::2,::2]
GREEN =CH3.variables['Rad'][:][::2,::2]
RED=CH2.variables['Rad'][:][::4,::4]
```

```python
RED=(RED-RED.min())/(RED.max()-RED.min())
GREEN=(GREEN-GREEN.min())/(GREEN.max()-GREEN.min())
BLUE=(BLUE-BLUE.min())/(BLUE.max()-BLUE.min())
```

```python
np.shape(RED)
```

```python
np.shape(GREEN)
```

```python
np.shape(BLUE)
```

```python
COLORES=np.zeros((5424, 5424,3))
```

```python
COLORES[:,:,0]=RED
```

```python
COLORES[:,:,1]=GREEN
```

```python
COLORES[:,:,2]=BLUE
```

```python
plt.figure(figsize = (5,5),dpi=150)
plt.imshow(COLORES)
```

```python
plt.figure(figsize = (5,5),dpi=150)
plt.xlim(2000,2500)
plt.ylim(2300,2000)
plt.title('RGB con canales 1-2-3 a las 19:30 UTC con el producto "raciances"')
plt.imshow(COLORES)
```
